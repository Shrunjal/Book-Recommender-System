POPULARITY BASED RECOMMENDER SYSTEM


```python
rating_with_book_name = ratings.merge(books,on="ISBN")
rating_with_book_name
```

```python
num_rating = rating_with_book_name.groupby("Book-Title")['Book-Rating'].count().reset_index()[["Book-Title","Book-Rating"]]
num_rating.rename(columns={"Book-Rating":"num_rating"},inplace=True)
```

```python
num_rating
```

```python
avg_rating = rating_with_book_name.groupby("Book-Title")['Book-Rating'].mean().reset_index()[["Book-Title","Book-Rating"]]
avg_rating.rename(columns={"Book-Rating":"avg_rating"},inplace=True)
avg_rating
```

```python
ratings = num_rating.merge(avg_rating,on="Book-Title")
ratings
```

```python
ratings = ratings[ratings["num_rating"]>=250].sort_values("avg_rating",ascending=False)
```

```python
merged_df = books.merge(ratings, on='Book-Title').drop_duplicates("Book-Title")
popular_df = merged_df[["Book-Title", "Book-Author", "Image-URL-M", "num_rating", "avg_rating"]]
```

```python
popular_df
```

```python
C = merged_df['avg_rating'].mean()
m = 100

# Filter books with more than m ratings
qualified = merged_df[merged_df['num_rating'] >= m].copy()

# Compute the weighted rating
qualified['score'] = (
    qualified['num_rating'] / (qualified['num_rating'] + m) * qualified['avg_rating']
    + m / (m + qualified['num_rating']) * C
)

# Sort by score
popular_df = qualified.sort_values('score', ascending=False)
popular_df = popular_df[["Book-Title", "Book-Author", "Image-URL-M", "num_rating", "avg_rating"]].head(50)

```

```python
popular_df
```

Collaborative Filter Based

```python
# x is the rating_with_book_name DataFrame
x = rating_with_book_name
x
```

```python
# Group by User-ID and count the ratings.
user_rating_counts = x.groupby("User-ID").count()["Book-Rating"]
user_rating_counts
```

```python
# Filter this Series to get the User-IDs where the count is greater than 200.
# This creates a new Series containing only the users who meet the criteria.
filtered_users = user_rating_counts[user_rating_counts > 200]
filtered_users
```

```python
# Get the index of this filtered Series, which contains the User-IDs we want.
UserID = filtered_users.index
UserID
```

```python
#Get number of total ratings done by 775 users
filtered_rating = rating_with_book_name[rating_with_book_name["User-ID"].isin(UserID)]
filtered_rating
```

```python
#Get ratings for each book
book_rating_counts = filtered_rating.groupby("Book-Title").count()["Book-Rating"]
book_rating_counts
```

```python
#Book ratings greater than 50
books_with_50_ratings = book_rating_counts[book_rating_counts >= 50]
books_with_50_ratings
```

```python
famous_books = books_with_50_ratings.index
famous_books
```

```python
final_rating = filtered_rating[filtered_rating["Book-Title"].isin(famous_books)].copy()
final_rating
```

```python
final_rating = final_rating.drop_duplicates()
final_rating
```

```python
table = final_rating.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
table
```

```python
table = table.fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(table)
```

```python
cosine_similarity(table).shape
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
  #index fetch
  index = np.where(table.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1],reverse = True)[1:6]

  data = []
  for i in similar_items:
    item = []
    temp_df = books[books["Book-Title"] == table.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
    data.append(item)
  return data
```

```python
recommend('The Da Vinci Code')
```

```python
recommend("A Painted House")
```

```python
recommend("The Firm")
```
